```python
# 1) Calculator with (addition, subtraction, multiplication, division, power)

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
op = input("Enter operator (+, -, *, /, **): ")

if op == "+":
    print("Result:", num1 + num2)
elif op == "-":
    print("Result:", num1 - num2)
elif op == "*":
    print("Result:", num1 * num2)
elif op == "/":
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error: Division by zero")
elif op == "**":
    print("Result:", num1 ** num2)
else:
    print("Invalid operator")
```

```python
# 2) Check if there is any numeric value in list help of using for loop

my_list = ["apple", "banana", 10, "orange"]
found = False

for item in my_list:
    if isinstance(item, (int, float)):
        found = True
        break

if found:
    print("Numeric value found in list")
else:
    print("No numeric value in list")
```

```python
# 3) Add a key to a dictionary

my_dict = {"name": "Ali", "age": 20}
my_dict["city"] = "Karachi"
print("Updated dictionary:", my_dict)
```

```python
# 4) Sum of all numeric items in a dictionary

my_dict = {"a": 10, "b": 20, "c": "hello", "d": 30}
total = 0

for value in my_dict.values():
    if isinstance(value, (int, float)):
        total += value

print("Sum of numeric values:", total)
```

```python
# 5) Identify duplicate values from list

my_list = [1, 2, 3, 2, 4, 5, 1, 6]
duplicates = []
seen = []

for item in my_list:
    if item in seen and item not in duplicates:
        duplicates.append(item)
    else:
        seen.append(item)

print("Duplicate values:", duplicates)
```

```python
# 6) Check if a given key already exists in a dictionary

my_dict = {"name": "Sara", "age": 22, "city": "Lahore"}
key = input("Enter key to check: ")

if key in my_dict:
    
    print("Key exists in dictionary")
else:
    print("Key does not exist in dictionary")
```
